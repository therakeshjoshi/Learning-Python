Basic Arithmetic

```python
a, b = 10, 5
```

```python
#Addition
print(f'a+b = {a+b}')
#Subtraction
print(f'a-b = {a-b}')
#Multiplication
print(f'a*b= {a*b}')
#Division 
print(f'a/b = {a/b}')
#Modulus
print(f'a%b = {a%b}')
#Exponentiation
print(f'a**b = {a**b}')
#Floor divison (will return largest whole no)
print(f'a//b = {a//b}')
#Bitwise AND
print(f'a&b= {a&b}')
#Bitwise OR
print(f'a|b = {a|b}')
#Booleans
print(f'a>b = {a>b}')
```
