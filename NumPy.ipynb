```python
#Integralss and Derivatives
```

```python
def f(x):
    return x**2
def derivative(f, x, h=1e-6):
    return (f(x+h)-f(x))/h
```

```python
a, b = 0 , 10
```

```python
x = np.linspace(a, b, 1000)
y = f(x)
```

```python
dy_dx = np.gradient(y, x)
```

```python
y_integral = np.cumsum(y)*(b-1)/len(x)
```

```python
integral_trapz = np.trapezoid(y, x)

y_integral[-1], integral_trapz
```

```python
#3x + 2y + z = 4
#5x - 5y + 4z = 3
#6x + z = 0
A = np.array([[3, 2, 1],[5,-5,4],[6,0,1]])
c = np.array([4,3,0])

np.linalg.solve(A,c)
```
